# CDS diagrams

## CDS Datasets

```python
import matplotlib.pyplot as plt
import os

# Data for the pie chart
labels = ['Climate indices', 'Climate projections', 'In-situ observations', 'Reanalysis', 
          'Satellite observations', 'Seasonal forecasts']
sizes = [3, 36, 14, 40, 37, 12]

# Calculate the total number
total_number = sum(sizes)

# Set default font properties for labels and titles
plt.rcParams.update({'font.size': 14, 'font.weight': 'bold'})

# Define color palette for the pie chart
colors = plt.get_cmap('Paired').colors

# Create the pie chart with absolute numbers and percentages
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(sizes, 
                                   labels=[f"{label} ({size})" for label, size in zip(labels, sizes)], 
                                   autopct='%1.1f%%', 
                                   startangle=140, 
                                   colors=colors,
                                   wedgeprops=dict(width=0.3))

# Customize font size and weight for labels and percentages
for text in texts:
    text.set_fontsize(12)
    text.set_fontweight('bold')
for autotext in autotexts:
    autotext.set_fontsize(12)
    autotext.set_fontweight('bold')

# Draw a smaller white circle in the center to create a "donut" style
centre_circle = plt.Circle((0, 0), 0.50, fc='white')
plt.gca().add_artist(centre_circle)

# Add the total number to the center of the diagram
plt.text(0, 0, f'Total\n{total_number}', ha='center', va='center', fontsize=20, fontweight='bold')

# Add a title to the plot
plt.title('C3S Datasets by Product Type', fontsize=18, fontweight='bold')

# Save the plot as an image file with improved file path handling
output_path = os.path.join('..', 'media', 'cds-datasets.png')
plt.savefig(output_path, format='png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
